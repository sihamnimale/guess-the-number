How to Play:
1. Run the `main()` function to start the game.
2. Enter your name to begin.
3. Choose a difficulty level: easy, medium or Hard.
4. Guess the hidden computer-generated number within the specified range.
5. Hints will be given to get you closer to the correct number.
6. Keep guessing until you find the correct number or run out of attempts.
7. Wins increase if you guess the correct number, losses increase if you run out of attempts.
8. Once game ends, choose to play again to improve your score.

```python
"""
Importing built-in random module -- to generate a random number within the specified range
"""
import random as rm

"""
Global variables to keep track of player's wins and losses
"""
wins = 0 
losses = 0

# Generate a random number between 1 and max_number depending on game level.
def random_number(max_number):
    n = rm.randint(1, max_number)
    return n
    

 
# Creating a display for the title of the game
def display_title():
    print('Guess The Number!')
    print()

    """
    display_title() function simply displays the title of the game 'Guess The Number
    """

# Creating input to ask for their name and greet the player
def greet_player():
    player_name = input('Please enter your name: ')
    print('Hello', player_name + '!')
    print()

    """
    greet_player() function prompts the player to enter their name 
    
    and greets them using the entered name
    """

# Creating a function to get input from the user and validating as a numeric value between 1 and max_number
def user_guess(max_number):
    
    """
    user_guess() function prompts the user to guess a number within a specified range. It repeatedly
    
    asks for input until the user enters a valid integer between 1 and max_number depending on the user's selected level
    
    The input is validated using a while loop and try-except block to handle invalid inputs
    """

    while True: 
        try: 
            guess = int(input('Guess a number between 1 and ' + str(max_number) + ': '))
            if 1 <= guess <= max_number:
                return guess 
            else: 
                print('Invalid input. Please enter within the specified range.')
        except ValueError: 
                print('Invalid input. Please enter a valid number.')


def validate_userguess(guess, n): 

    """
    validate_userguess() function validates the user's guess and provides 
    
    feedback based on whether the guess is too high, too low, or correct. The function 
    
    takes two parameters the user's guessed number(guess) and the correct number to be guessed (n). 
    
    It is integrated to inform the user about the accuracy of their guess
    """

    if guess > n:
        print('Your number is too high. Try Again.')
        return False #to indicate that the guess is incorrect
    elif guess < n:
        print('Your number is too low. Try Again.')
        return False
    else: 
        print('Correct, your number is a match!')
        return True #to indicate the guess is correct
            
# Main game function that arranges the number-guessing game
def game():
    global wins, losses # used to track player's performance in the game

    """
    Allows the user the input their chosen game level: 
    
    easy (1 to 10); medium (1 to 100); or hard (1 to 1000). It continues to 
    
    prompt until a valid input is provided
    """

    while True:
        level = input('Input your chosen game level. Easy, Medium or Hard: ')
        if level.lower() in ['easy', 'medium', 'hard']:
            break
        else:
            print('Invalid input. Please try again.')
            
    """
    Set the max_number and number of chances  the player has to guess the correct number
    
    based on the selected game level
    """

    if level.lower() == 'easy':
        max_number = 10
        chances = 3
    elif level.lower() == 'medium':
        max_number = 100
        chances = 5
    elif level.lower() == 'hard':
        max_number = 1000
        chances = 10

# Generating the winning number within the specified range
    winning_number = random_number(max_number)
    attempts = chances 

    """
    Game loop to remind the user how many attempts they have left to guess the number. 
    
    Feedback is provided based on whether the guess is too high, too low, or correct. The loop 
    
    continues until the player runs out of attempts (attempts = 0)
    """

    while attempts > 0: 
        print(f'You have {attempts} attempts remaining.')
        userguess = user_guess(max_number) 
        is_user_correct = validate_userguess(userguess, winning_number)
    
        """
        Generate an update of total wins and losses when there are no attempts remaining and the game is over.
            
        If the player guesses the correct number within the given attempts, the wins variable is incremented. 
            
        Otherwise, the losses variable is incremented and the correct number is revealed
        """
        if is_user_correct:
            wins += 1
            return
        else: 
            attempts -= 1
        if attempts == 0:
            losses += 1
            print(f'Game over! The number is {winning_number}.')
            return
                     
# Main function to start the game 
def main(): 
    global wins, losses #track player's performance
    
    display_title()
    greet_player()

    """
    The function enters a loop where the game() function is repeatedly called, 
    
    allowing the player to play the game multiple times if desired
    """

    while True:
        game()

# Determining the pluralisation of 'game' when total count is over 1 for  total win and loss
        w_text = 'game' if wins == 1 else 'games'
        l_text = 'game' if losses == 1 else 'games'
        print(f'You have won {wins} {w_text} and lost {losses} {l_text}.') #Display the player's wins and losses

# End of the game to ask if the player would like to play again
        while True:
            again = input('Would you like to play again? (y/n): ')
            if again.lower() == 'y':
                break
            elif again.lower() == 'n':
                print('Goodbye!')
                return
            else: 
                print('Invalid input. Please enter y or n.')
        print()

# In order to run the game
if __name__ == "__main__":
    main()
```
